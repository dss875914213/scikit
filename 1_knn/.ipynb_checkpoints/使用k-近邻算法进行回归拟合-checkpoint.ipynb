# 生成数据集

```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = make_regression( n_features=1, noise=7)
```

```python
data[0].shape
```

```python
X = data[0]
y = data[1]
```

```python
plt.figure(figsize=(4,3),dpi=144)
plt.scatter(data[0],data[1])
```

# 训练

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
KNN = KNeighborsRegressor()
```

```python
KNN.fit(X,y)
```

```python
KNN.score(X,y)
```

```python
print(X.min(),X.max())
```

# 针对足够点进行预测

```python
predict_X = np.arange(X.min(),X.max(),(X.max()-X.min())/100.0)[:,np.newaxis]
```

```python
KNN.predict([[1]])
```

# 将这些点连起来

```python
plt.figure(figsize=(4,3),dpi=144)
predict_y = KNN.predict(predict_X)
plt.scatter(X,y,s=5,c='r')
plt.plot(predict_X,predict_y)
```
